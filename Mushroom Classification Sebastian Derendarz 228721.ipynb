### d) Wybór cech oraz przekształcenia danych

Zdecydowałem się na dokonanie trzech klasyfikacji:
- z wykorzystaniem jednego atrybutu z tych, które w raporcie z profilowania danych zostały wskazane jako wysoko skorelowane z kolumną wynikową. DataFrame zostanie nazwany `df2`.
- z wykorzystaniem wszystkich atrybutów, które w raporcie z profilowania danych zostały wskazane jako wysoko skorelowane z kolumną wynikową. DataFrame zostanie nazwany `df3`.
- z wykorzystaniem wszystkich atrybutów. DataFrame zostanie nazwany `df4`.  

Wartości tekstowe w powstałych DataFrame-ach zostaną skonwertowane do wartości liczbowych przy wykorzystaniu funkcji `get_dummies` z biblioteki `pandas`. Bazowy DataFrame - `df` nie ulegnie zmianie.

```python
def keep_cols(dataframe, keep_these):
    """Keep only the columns [keep_these] in a DataFrame, delete
    all other columns. 
    """
    drop_these = list(set(list(dataframe)) - set(keep_these))
    return dataframe.drop(drop_these, axis = 1)

df2 = df.pipe(keep_cols, ['edibility', 'odor']).pipe(pd.get_dummies, drop_first=True)

df3 = df.drop(
        columns=[
            'cap shape',
            'cap surface',
            'cap color',
            'gill attachment',
            'stalk shape',
            'veil color',
            'ring number',
            'ring type',
            'habitat'
        ]
    ).pipe(pd.get_dummies, drop_first=True)

df4 = pd.get_dummies(df, drop_first=True)
```

# 5. Tworzenie modelu oraz ocena trafności predykcji

### a) Funkcja służąca do predykcji jadalności grzyba

Kompaktowa, implementująca algorytm _k-nearest neighbos_ (k najbliższych sąsiadów). Zostanie wykorzystana do analizy datasetu niezależnie od cech, które wybraliśmy. Proszę zapoznać się z docstringiem oraz kodem w celu zrozumienia sposobu działania oraz wykorzystania funkcji. Wartości zwrócone przez funckję mogą być w prosty sposób ocenione poprzez wywołanie funkcji `accuracy_score(yk_test, yk_pred)` z biblioteki `sklearn.metrics`. Wynik wywołania tej funkcji będę uznawał za ostateczny wyznacznik jakości klasyfikacji.

```python
from typing import Tuple

import numpy as np
from sklearn import neighbors
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

def predict_edibility(dataframe: pd.DataFrame) -> Tuple[np.ndarray, np.ndarray]:
    '''Predict mushroom edibility.
    
    One condition that must be met is that desired result of classification
    must be stored in the first column of DataFrame in a column named 'edibility_POISONOUS'.
    
    params:
        dataframe: pandas DataFrame object to be analized
    returns:
        Tuple of test and predicted values for dataframe
        than can be used to assess accuracy of prediction.
    '''
    X = dataframe.iloc[:, 1:]
    y = dataframe['edibility_POISONOUS']
    yk = y.values

    # division into test and teach sets of values
    idx_ucz, idx_test = train_test_split(np.arange(X.shape[0]),test_size=0.9, random_state=12345)

    X_ucz, X_test = X.iloc[idx_ucz, :], X.iloc[idx_test, :]
    y_ucz, y_test = y[idx_ucz], y[idx_test]
    yk_ucz, yk_test = yk[idx_ucz], yk[idx_test]
    X_ucz.shape, X_test.shape, y_ucz.shape, y_test.shape

    knn = neighbors.KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_ucz, yk_ucz)
    yk_pred = knn.predict(X_test)
    return yk_test, yk_pred
```

### b) Ocena trafności predykcji modelu wykorzytującego jedynie jeden z najwyżej skorelowanych atrybutów

```python
yk_test, yk_pred = predict_edibility(df2)
accuracy = accuracy_score(yk_test, yk_pred)


print(f'Accuracy of prediction: {accuracy*100:.2f}%')
```

### c) Ocena trafności predykcji modelu wykorzytującego jedynie najwyżej skorelowane atrybuty

```python
yk_test, yk_pred = predict_edibility(df3)
accuracy = accuracy_score(yk_test, yk_pred)


print(f'Accuracy of prediction: {accuracy*100:.2f}%')
```

### d) Ocena trafności predykcji modelu wykorzytującego wszystkie atrybuty

```python
yk_test, yk_pred = predict_edibility(df4)
accuracy = accuracy_score(yk_test, yk_pred)

print(f'Accuracy of prediction: {accuracy*100:.2f}%')
```
